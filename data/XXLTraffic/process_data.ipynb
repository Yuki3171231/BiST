```python
import csv
import pandas as pd
import numpy as np
```

```python
day = pd.read_csv('data/XXLTraffic/pems05_d.csv')
day = day.fillna(0)
day = day.drop(day.columns[0], axis=1)
time_index = pd.date_range(start='2012-03-07', 
                            periods=day.shape[0], 
                            freq='d')
day = pd.DataFrame(day.values, 
                   index=time_index, 
                            )
print(day.head())
day.to_hdf('data/XXLTraffic/his_day.h5', key='t', mode='w')
```

```python
hour = pd.read_csv('data/XXLTraffic/pems05_h.csv')
hour = hour.fillna(0)
hour = hour.drop(hour.columns[0], axis=1)
time_index = pd.date_range(start='2012-03-07', 
                            periods=hour.shape[0], 
                            freq='h')
hour = pd.DataFrame(hour.values, 
                   index=time_index, 
                            )
print(hour)
hour.to_hdf('data/XXLTraffic/his_hour.h5', key='t', mode='w')
```

```python
common = pd.read_csv('data/XXLTraffic/pems05_all_common_flow.csv')
common = common.fillna(0)
common = common.drop(common.columns[0], axis=1)
time_index = pd.date_range(start='2012-03-07', 
                            periods=common.shape[0], 
                            freq='5T')
common = pd.DataFrame(common.values, 
                   index=time_index, 
                            )
# 重新采样到每15分钟
common = common.resample('15T').mean().round(0)

# 填充缺失值为0
common = common.fillna(0)

# print(common)
common.to_hdf('data/XXLTraffic/his_common.h5', key='t', mode='w')
```
